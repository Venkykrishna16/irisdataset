```python
#import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the iris dataset
url = "/content/iris.csv"
iris_data = pd.read_csv(url)
```

```python
#Display the first few rows of the datset
print("first 5 rows of the iris dataset:\n")
print(iris_data.head())
```

```python
# Display summary statistics of the dataset
print("\nBasic information about the dataset:\n")
print(iris_data.info())
```

```python
# Display summary statistics of the dataset
print("\nSummary Statistics:\n")
print(iris_data.describe())
```

```python
# Check for missing values
print("\nMissing values in each column:\n")
print(iris_data.isnull().sum())
```

```python
# Data visualisation

#pairplot
sns.pairplot(iris_data,hue='species')
plt.show()
```

```python
#3D scatter plot
iris_data = pd.read_csv('/content/iris.csv')
iris_data.head()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(iris_data['sepal_length'],
           iris_data['sepal_width'],
           iris_data['petal_length'],
           c=iris_data['species'].astype('category').cat.codes,
           cmap='viridis')

ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')


plt.show()
```

```python
#4D scatter plot
iris_data['species_num'] = iris_data['species'].astype('category').cat.codes
plt.scatter(iris_data['sepal_length'], iris_data['sepal_width'], c=iris_data['species_num'], s=iris_data['petal_length']*10, alpha=0.5, cmap='viridis')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.colorbar(label='species')
plt.show()
```
